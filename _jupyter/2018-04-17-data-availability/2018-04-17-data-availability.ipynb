This notebook is the scratch work behind my post on microbiome data availability.

```python
import pandas as pd
```

```python
fname = 'Disease datasets list - data blog post.tsv'
df = pd.read_csv(fname, sep='\t')
df.head()
```

```python
# All rows in "data availability" column are filled out
df.shape, df.dropna(subset=['Data availability']).shape
```

```python
df['full_data'] = ['AWS' in str(i) for i in df['Status'].values]
print(len([i for i in df['Status'].dropna().values if 'AWS' in i]))
```

```python
#df.dropna(subset=['Manual meta analysis?'])[['dataset_id', 'Manual meta analysis?', 'Data availability', 'Status']]
df.groupby('full_data').size()
```

```python
df['Data availability'] = df['Data availability'].str.lower()
df.groupby(['full_data', 'Data availability']).size()
```

# Have full data

```python
df.query('full_data == True').sort_values(by='Data availability')
```

```python
internet = ['ena', 'sra', 'mg-rast', 'author website']
df = df.rename(columns={'Data availability': 'data_avail'})
df.query('data_avail == @internet').sort_values(by='Clinical metadata availability')
```

## Have full data, got data from internet

```python
df.query('full_data == True')\
    .query('data_avail == @internet')\
    .groupby(['data_avail', 'Clinical metadata availability'])\
    .size()
```

```python
df.query('full_data == True').query('data_avail == @internet').shape
```

## Have full data, didn't get data from internet

```python
df.query('full_data == True')\
    .query('data_avail != @internet')\
    .groupby(['data_avail', 'Clinical metadata availability'])\
    .size()
```

# Don't have full data

```python
df.query('full_data == False').groupby('data_avail').size().sort_values()
```

## Don't have full data, but raw data on SRA

```python
df.query('full_data == False').query('data_avail == "sra"')
```

```python
df.query('full_data == False').query('data_avail == "author website"')
```

## Don't have full data, emailed

```python
df['data_avail'] = df['data_avail'].replace('would need to email', 'emailed')
df.query('full_data == False').query('data_avail == "emailed"').shape
```

```python
df.query('full_data == False').query('data_avail == "emailed"').groupby("If emailed, response").size()
```

```python
got_data = ['Got data from authors', 'Responded, sent via ftp']
df = df.rename(columns={"If emailed, response": 'email_response'})

keepcols = ['Manual meta analysis?', 'data_avail', 'email_response', 
            'Comments on email exchange/data', 'Clinical metadata availability',
            'Clinical metadata notes', 'Processing information',
            'Other comments', u'Status', u'Name in AWS', u'Email sent out?',
            'Disease', 'First and last authors']

df.query('full_data == False').query('data_avail == "emailed"').query('email_response == @got_data')[keepcols]
```
